```

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/mnt/data/healthcare_dataset.csv')

# 1. Age Distribution
plt.figure()
plt.hist(df['Age'].dropna())
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.tight_layout()
plt.show()

# 2. Billing Amount Distribution
plt.figure()
plt.hist(df['Billing Amount'].dropna())
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.title('Billing Amount Distribution')
plt.tight_layout()
plt.show()

# 3. Room Number Distribution
plt.figure()
plt.hist(df['Room Number'].dropna())
plt.xlabel('Room Number')
plt.ylabel('Frequency')
plt.title('Room Number Distribution')
plt.tight_layout()
plt.show()

# 4. Medical Condition Frequency
plt.figure()
df['Medical Condition'].value_counts().plot(kind='bar')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.title('Frequency of Medical Conditions')
plt.tight_layout()
plt.show()

# 5. Admission Type Frequency
plt.figure()
df['Admission Type'].value_counts().plot(kind='bar')
plt.xlabel('Admission Type')
plt.ylabel('Count')
plt.title('Frequency of Admission Types')
plt.tight_layout()
plt.show()

# 6. Medication Frequency
plt.figure()
df['Medication'].value_counts().plot(kind='bar')
plt.xlabel('Medication')
plt.ylabel('Count')
plt.title('Frequency of Medication')
plt.tight_layout()
plt.show()
```
